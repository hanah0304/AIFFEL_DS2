# 택시 회사 매출 증진 방안
1. 승객수 당 평균 운임 확인
2. 결제 방법에 따른 매출 차이
3. 이동 거리 대비 운임 분석
4. 승차 시간대별 수익 분석
5. 시간대별 팁 금액 차이
6. 재탑승에 따른 운임 변화 추이
7. 시간x요일에 따른 수요 변화

```python
from datetime import datetime

# 승객cnt당 평균 운임 및 팁 계산
data['total_amount'] = data['fare_amount'] + data['tip_amount'] + data['tolls_amount']
average_fare_per_passenger = data.groupby('passenger_count')['total_amount'].mean().reset_index()
print(average_fare_per_passenger)
```

```python
# 결제 방법에 따른 매출 차이
payment_method_revenue = data.groupby('payment_method')['total_amount'].sum().reset_index()
print(payment_method_revenue)

# 카드 자동 결제 설정 시, 다음 탑승 시 사용 가능한 쿠폰 발급 프로모션 제안
```

```python
# 이동 거리 대비 운임
data['fare_per_mile'] = data['fare_amount'] / data['trip_distance']
fare_per_mile_analysis = data.groupby('trip_distance')['fare_per_mile'].mean().reset_index()
print(fare_per_mile_analysis)
```

```python
# 승차 시간대별 운임 차이
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])
data['pickup_hour'] = data['tpep_pickup_datetime'].dt.hour

# 시간대별 매출 분석
hourly_revenue = data.groupby('pickup_hour')['total_amount'].sum().reset_index()
print(hourly_revenue)
```

```python
# 탑승 시간, 이동 거리, 운임 사이의 관계 확인

# 히트맵을 그리기 위해 변수들 간의 상관관계 계산
corr = data[['trip_distance', 'fare_amount', 'tip_amount']].corr()

# 히트맵 그리기
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', annot_kws={"size": 12})
plt.title('Correlation Heatmap of Trip Distance, Fare Amount, and Tip Amount')
plt.show()
```

```python
# 산점도 그리기
plt.figure(figsize=(10, 6))
plt.scatter(data['trip_distance'], data['fare_amount'], s=100, alpha=0.7, edgecolors='w')

# 축과 제목 설정
plt.title('Fare Amount vs. Trip Distance')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Fare Amount ($)')

# 그리드 추가
plt.grid(True)

# 그래프 출력
plt.show()
```

```python
# 시간대별 팁 금액 평균 계산
tip_by_hour = data.groupby('pickup_hour')['tip_amount'].mean().reset_index()

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x='pickup_hour', y='tip_amount', data=tip_by_hour, palette='viridis')
plt.title('Average Tip Amount by Pickup Hour')
plt.xlabel('Pickup Hour')
plt.ylabel('Average Tip Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 승객 이름을 기준으로 그룹화하여 평균 운임과 등장 횟수 계산
grouped_data = data.groupby('passenger_name').agg({
    'fare_amount': 'mean',
    'passenger_name': 'size'
}).rename(columns={'passenger_name': 'count'}).reset_index()

# 등장 횟수(count) 기준으로 내림차순 정렬
grouped_data_sorted = grouped_data.sort_values(by='count', ascending=False)

# 결과 출력
print(grouped_data_sorted)
```

```python
# 산점도 시각화
plt.figure(figsize=(10, 6))
plt.scatter(grouped_data['count'], grouped_data['fare_amount'], alpha=0.7, color='b', edgecolors='k')
plt.title('Relationship between Number of Appearances and Average Fare Amount')
plt.xlabel('Number of Appearances')
plt.ylabel('Average Fare Amount')
plt.grid(True)
plt.show()
```

```python
# 시간대별 수요량 분석

# 시간대별 승객 수 계산
hourly_demand = data.groupby('pickup_hour')['passenger_count'].sum().reset_index()
hourly_demand = hourly_demand.sort_values(by='passenger_count', ascending=False)

# 결과 출력
print(hourly_demand)
```

```python
# 승차일 요일 추출
data['pickup_day_of_week'] = data['tpep_pickup_datetime'].dt.day_name()

# 요일별 승객 수 계산
weekday_demand = data.groupby('pickup_day_of_week')['passenger_count'].sum().reset_index()

# 요일 순서대로 정렬
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_demand['pickup_day_of_week'] = pd.Categorical(weekday_demand['pickup_day_of_week'], categories=weekday_order, ordered=True)

# 요일별 승객수를 내림차순으로 정렬
weekday_demand = weekday_demand.sort_values('passenger_count', ascending=False)

# 결과 출력
print(weekday_demand)
```

```python
# 요일x시간대별 승객수 비교
# 요일과 시간대별 승객 수 합계 계산
heatmap_data = data.groupby(['pickup_day_of_week', 'pickup_hour'])['passenger_count'].sum().unstack()

# 요일 순서 설정
heatmap_data = heatmap_data.reindex(weekday_order)

# 히트맵 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Passenger Count by Day of Week and Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.show()
```
